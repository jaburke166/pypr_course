```python
# Question 1 

def common(x,y):
    '''
    This will take two lists and return a True if they contain elements in
    common. 
    '''
    return len(set(x).intersection(set(y))) > 0
```

```python
# Question 2

def diffs(x,y):
    '''
    Takes two lists as arguments and returns the differences between them
    '''
    union=list(set(x).union(set(y)))
    intersection = list(set(x).intersection(set(y)))
    for i in range(len(intersection)):
        union.remove(intersection[i])
    return sorted(union)
```

```python
# Question 4

import numpy as np
def threed_angle(a,b):
    '''
    Takes two vectors given in length 3 Numpy arrays and determines the angle
    between them
    '''
    theta = (180/np.pi)*np.arccos(np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b)))
    if theta > 0 and theta < 180:
        return theta
    elif theta > 180:
        return 360-theta
    else:
        return -1*theta
```

```python
# Question 5 

import numpy as np

def med_hw(x):
    '''
    Recieves a list of tuples in form (height, weight) and outputs the median
    height and median weight
    '''
    heights, weights = zip(*x)
    return (np.median(heights), np.median(weights))
```

```python
# Question 6

def dict_build(keys, values):
    '''
    This takes an argument of two lists of keys and values and returns a 
    dictionary 
    '''
    if len(keys)!=len(values): #If the lists do not have equal length return None
        return None
    else:
        tuple_list = zip(keys,values) #Zip together the two lists otherwise and return dictionary
        return dict(tuple_list)
```

```python
# Question 7

def cof_tea(x):
    '''
    This takes a dictionary as its argument and splits it into two lists
    containing the names of whether they like coffee or tea
    '''
    a=[]
    b=[]
    for key, value in x.items():
        if value == 'tea':
            a.append(key)
        else:
            b.append(key)
    return sorted(b),sorted(a)
```

```python
# Question 8

def string_reader(x):
    '''
    This accepts the input of a string and returns a dictionary with each letter
    in the string as a key and the number of time that letter has appeared as
    the value
    '''
    mydict={}
    splitstring = x.split() #Split string into individual words without the space character
    for word in splitstring: #Loops over the words in the splitted string
        for letter in word: #Loops over the letters in the word
            if letter in mydict.keys(): #This determines whether the letter already exists in the dictionary or not
                mydict[letter]+=1 #If it does, increment the number of times it's appeared
            else:
                mydict[letter] = 1 #Otherwise, initialise its appearance as 1
    return mydict
```

```python
# Question 9

def student_res():
    '''
    Classify student results from the given dict in the way 
    instructed.
    '''
    scores = {'s151': 41,'s152': 55,'s153': 82,'s154': 47,'s155': 35,'s156': 34,'s157': 64,'s158': 57,
          's159': 73, 's161': 82, 's162': 93, 's163': 54, 's164': 85, 's165': 16, 's166': 24}
    # Insert your code here.
    first = []
    second = []
    for key,val in scores.items(): #Split dictionary into first and second years
        if key[:3]=='s15':
            second.append(val)
        else:
            first.append(val)
    #Determine the number of students that passed in each year
    second_pass = sum([second[i] > 40 for i in range(len(second))])
    first_pass = sum([first[i] > 40 for i in range(len(first))])

    #Determine percentages of who passed in first, second and in both years
    secondpct = 100*second_pass/len(second)
    firstpct = 100*first_pass/len(first)
    overallpct = 100*(second_pass+first_pass)/(len(first)+len(second))
    return round(overallpct,2), round(firstpct,2), round(secondpct,2)
```

```python
# Question 10
import numpy as np

# This function actually constructs a chess board and then checks which colour
def chess(x):
    '''
    This function will return the colour of the square positioned at x on a 
    8x8 chessboard
    '''
    letters=['a','b','c','d','e','f','g','h'] #List the column letters
    col,row = x[0],x[1] #Split up the code for the place on the board

    col_n = letters.index(col) #Determine the column number via the index in the
    #letters list
    row_n = 8-int(row) #Since arrays work by going down and right whereas
    #chessboards move up and right we invert the row number by subtracting it
    #from 8

    board = np.zeros((8,8)) #This sets up the board with 1s and 0s in the 
    #criss-cross pattern that a chessboard is shaped in. 1s represent white and
    #0s represent black
    for i in range(board.shape[0]):
        for j in range(board.shape[1]):
            if i%2==0 and j%2==0: #If row and column both even
                board[i,j]=1
            elif i%2!=0 and j%2!=0: #If row and column both odd
                board[i,j]=1
    if board[row_n,col_n]==1: #If 1 then white and vice versa
        return 'white'
    else:
        return 'black'
    
# This function is alot simpler and to the point
def chess_simpler(x):
    '''
        This function takes input of a square on a chess board,
        and returns whether the square is black or white.
    '''
    if x[0] in 'aceg':
        if int(x[1])%2==0:
            return 'white'
        else:
            return 'black'
    else:
        if int(x[1])%2==0:
            return 'black'
        else:
            return 'white'
       
```
