```python
# Question 1

def sum_check(n):
    '''
    Checks if the sum to n of integers i and i^2 match the 
    usual summation formula, returns True if they do, False if one
    does not.
    '''
    # Initialise the sum totals
    total_1 = 0
    total_2 = 0
    
    #Start loop to determine iterative summation
    for i in range(0,n+1):
        total_1 += i
        total_2 += i**2
    
    #Calculate the value without iteration
    check_1 = (n*(n+1))/2
    check_2 = (n*(n+1)*(2*n+1))/6
    
    #Write to boolean variables on whether the two answers agree or not
    if check_1 == total_1 and check_2 == total_2:
        check = True
    elif  check_1 != total_1 or check_2 != total_2:
        check = False
    return check
```

```python
# Question 3

def leap_year(n):
    '''
        This function uses if statements to determine whether and inputed year
        is a leap year or not, returning a boolean with the answer.
    '''
    #Initialise the answer as false
    f = False
    # Create IF condition satisfying leap year 
    if n%4 == 0 and n%100!=0 or n%4==0 and n%400==0:
        f=True
    return f   
```

```python
# Question 4

def seasons(d,m):
    '''
    Function returns a string with the name of the season for a
    given day an month input as integer arguments
    '''
    if m<=2 or (m==12 and d>=21) or (m==3 and d<=20):
        returnValue = 'Winter'
    elif m<=5 or (m==6 and d<=20):
        returnValue = 'Spring'
    elif m<=8 or (m==9 and d<=20):
        returnValue = 'Summer'
    else:
        returnValue = 'Autumn'
    return returnValue 
```

```python
# Question 5

def factors(n):
    '''
        This function creates a list of all of the factors of a given number n.
    '''
    factor_list=[]
    if n%2 == 0: # Check to see if n is even
        for i in range(2, int(n/2)+1): #Run loop from 2 to n/2
            if n%i == 0:
                factor_list.append(i) #Append to list if 0 remainder
    else:
        for i in range(2,int((n+1)/2)): #Similar to above but when n is odd
            if n%i == 0:
                factor_list.append(i)
    return factor_list
```

```python
# Question 7
def highest_integer():
    '''
        This function finds the highest integer n such that
        n^2+6n-3<3000
        and the smallest integer m such that 
        m^2+6m-3>500
    '''
    n=0
    m=20  
    while n**2 + 6*n - 3 < 3000:
        n += 1
        if m**2 + 6*m -3 > 500:
            m -= 1
    n -= 1
    m += 1
    return m,n
```

```python
# Question 8
def evens():
    '''
    Generator to produce even numbers
    2,4,6,8,...
    '''
    i = 2
    while True:
        yield i
        i+=2
```

```python
# Question 9
import numpy as np

def even_sum():
    '''
    Generator for the Fibonacci sequence.
    '''
    # Using the fibonacci generator from the workshop.
    def fib_gen():
        a = 1
        b = 2
        while True:
            yield a
            a,b = b,a+b
    
    #Generate the fibonnaci list until threshold is reached.
    fib = fib_gen()
    fib_list = [1]
    while fib_list[-1] < 4e6:
        fib_list.append(next(fib))
        
    #List comprehension to store the even values from the fibonnaci list
    fib_even = [x for x in fib_list if x%2==0]
    
    #Calculate and print sum
    sum = np.sum(fib_even)
    print("The sum of the even fibonacci numbers below 4,000,000 is " + str(sum) + ".")
```

```python
# Question 10

def primes():
    '''
    Function uses the sieve of Eratosthenes method to return a list
    of primes <=1000.
    '''
    #Initialise set of all integers from 2-999
    primes = [x for x in range(2,1000)]
    
    # If the remainder of any division between any of the elements in the list
    # is 0 then it is removed from the list.
    for y in primes:
        for x in primes:
            if x%y == 0 and x!=y:
                primes.remove(x)
    return primes 
```
