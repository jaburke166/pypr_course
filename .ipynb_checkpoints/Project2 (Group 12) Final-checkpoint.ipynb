```python
# load required packages and read in provided data
import pandas as pd
import sys
import os
from collections import namedtuple
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import folium
import time as t
# Install word cloud presentation kit
import sys
!{sys.executable} -m pip install wordcloud

import wordcloud 
```

```python
class ReviewsAnalyzer:
    """
    Build an emotional analysis class of comments. 
    Load both positive and negative dictionaries as matching criteria, giving an emotional score to each comment and title . 
    Keep the score between [-1, 1], and the higher the score, the more positive it is.
    A score of 0 is neutral
    """
    def __init__(self, pos_f, neg_f):
        self.pos_dict = self.dictionary_load(pos_f)
        self.neg_dict = self.dictionary_load(neg_f)

    def reviews_score(self, sentence):
        """
        Grade a sentence.
        Score = (number of positive words - number of negative words)/total number of words of the sentences
        """
        from re import split
        words = [word for word in split("[^a-z-]+", sentence.lower()) if word]
        #Extract positive words
        pos_cnt = len([word for word in words if word in self.pos_dict])
        #Extract negative words
        neg_cnt = len([word for word in words if word in self.neg_dict])
        #Grade a sentence
        score = (pos_cnt - neg_cnt) / len(words) if len(words) else .0
        return score
    
    def reviews_score_update(self, sentence):
        """
        Grade a sentence.
        Score = (number of positive words - number of negative words)/total number of words of the sentences.
        If a negative word appears and it's followed by an emotional word, the score of subsequent emotional word is multiplied by -1
        """
        from re import split
        neg_words = ['not', 'no']
        words = [word for word in split("[^a-z-]+", sentence.lower()) if word]
        s_lst = [1 if w in self.pos_dict else -1 if w in self.neg_dict else 0
                 for w in words]
        for i, s in enumerate(s_lst[:-1]):
            if (words[i] in neg_words) and s_lst[i+1]!=0:
                s_lst[i], s_lst[i+1] = 0, -s_lst[i+1]
        score = sum(s_lst) / len(words) if len(words) else .0
        return score

    @staticmethod
    def dictionary_load(dict_f):
        """
        Loading the word from the dictionary file and returning the vocabulary set
        """
        with open(dict_f) as f:
            lines = f.readlines()
        dict_ = set([line.strip() for line in lines if not line.startswith(";")])
        if '' in dict_:
            dict_.remove('')
        return dict_
```

```python
# Initialize an emotion analysis class with a word file
# positive dictionary
POS_FILE = "positive-words.txt"
# negative dictionary
NEG_FILE = "negative-words.txt"  
ra = ReviewsAnalyzer(POS_FILE, NEG_FILE)

# Loading comment data
REV_FILE = "WomensApparelReviews.csv"
review_df = pd.read_csv(REV_FILE)

# Score each title with a defined emotional analysis class
title_score = [ra.reviews_score(sent) if isinstance(sent, str) else .0 
               for sent in review_df["Title"].values ]
review_df['title_score'] = title_score

# Score each review text with a defined emotional analysis class
review_score = [ra.reviews_score(sent) if isinstance(sent, str) else .0 
               for sent in review_df["Review Text"].values ]
review_df['review_score'] = review_score
```

```python
# all functions defined to answer the questions in this task are collected inside the Class of functions Question2. Because
# this class expects only 1 input of a dataframe, the class will need to be called each time a dataframe is to be analysed.

class Question2:
    
    def __init__(self, df):
        '''The initialisation function expects one arguments to initialise which dataframe the functions are being applied to
        
        df: DataFrame to be analysed'''
        self.df = df
        
    def trips_week(self):
        '''This will return the number of trips that occurred on every day of the week'''
        
        self.df['dow'] = self.df['started_at'].dt.dayofweek.values
        trips_dow = self.df.groupby('dow').count()['started_at']
        days = trips_dow.index
        return (days, trips_dow)

    def trips_hour(self):
        '''This will return the number of trips that occurred on every hour of the day'''

        self.df['hr'] = self.df['started_at'].dt.hour.values
        trips_hr = self.df.groupby('hr').count()['started_at']
        hrs = trips_hr.index
        return (hrs, trips_hr)

    def spatial_coords(self, idx):
        '''This will return the start and end coordinates for the idx'th trip of the Dataframe.

        idx: index/trip on dataframe to extract locations from'''

        x1 = self.df['start_station_latitude'][idx]
        x2 = self.df['end_station_latitude'][idx]
        y1 = self.df['start_station_longitude'][idx]
        y2 = self.df['end_station_latitude'][idx]
        return (x1, y1), (x2, y2)

    def distance(self, P1, P2):
        '''This will return the distance between two 2D points.

        P1, P2: 2-tuple floats representing coordinates of stations'''

        return np.linalg.norm(P1-P2)

    def hav_distance(self, P1,P2):
        '''computes the Haversine distance which determines the great circle distance between two coordinates of longitude
        and latitude. 

        P1, P2: 2-tuple floats representing coordinates of stations'''
        
        # Convert coordinates to radians
        lat1,long1 = np.radians(P1[0]),np.radians(P1[1])
        lat2,long2 = np.radians(P2[0]),np.radians(P2[1])
        
        # radius of Earth in Km 
        r = 6371 
        
        # use formula for Haversine distance from Wikipedia 
        hav = np.sin((lat2-lat1)/2)**2 +np.cos(lat1)*np.cos(lat2)*np.sin((long2-long1)/2)**2
        dist = 2*r*np.arcsin(np.sqrt(hav))
        
        return dist


    def trips_startend_station(self, station):
        '''This will return the trips that started at 'station' and the trips that ended at 'station'

        station: Station name'''

        start = self.df.loc[self.df['start_station_name'].str.startswith(station)]
        end = self.df.loc[self.df['end_station_name'].str.startswith(station)]
        
        return start, end

    def count_starts(self, coord):
        '''This will return the number of round trips that occured at the station coordinates coord. A round trip is defined
        as a trip where the start and end station are the same.

        coord: 2-tuple of latitude and longitude of station'''

        num = len(self.df.loc[(round(self.df['start_station_latitude'],3) == coord[0]) &\
                              (round(self.df['start_station_longitude'],3) == coord[1])])
        return num

    
    def get_arrows(self, locations, color, size=10, n_arrows=3):
        '''
        NOTE: This function was taken directly from a Medium blog post found at: 
            https://medium.com/@bobhaffner/folium-lines-with-arrows-25a0fe88e4e
        And as such, we take no credit for this work and thank Bob Haffner for providing the code.
        
        
        Get a list of correctly placed and rotated arrows/markers to be plotted

        locations : list of lists of lat lons that represent the 
                    start and end of the line. 
                    eg [[41.1132, -96.1993],[41.3810, -95.8021]]
                    
        arrow_color : default is 'blue'
        
        size : default is 10
        
        n_arrows : number of arrows to create.  default is 3
        
        Return
        list of arrows/markers
        
        Notes
        Based on https://gist.github.com/jeromer/2005586
        '''

        # Point named tuple
        Point = namedtuple('Point', field_names=['lat', 'lon'])

        # creating point from our Point named tuple
        p1 = Point(locations[0][0], locations[0][1])
        p2 = Point(locations[1][0], locations[1][1])        

        # Convert coordinates to radians and determine difference in longitudes to determine compass bearing below
        long_diff = np.radians(p2.lon - p1.lon)
        lat1 = np.radians(p1.lat)
        lat2 = np.radians(p2.lat)

        #Compute the compass bearings between points p1 and p2
        x = np.sin(long_diff) * np.cos(lat2)
        y = (np.cos(lat1) * np.sin(lat2) 
            - (np.sin(lat1) * np.cos(lat2) 
            * np.cos(long_diff)))
        bearing = np.degrees(np.arctan2(x, y))

        # adjusting for compass bearing
        if bearing < 0:
            bearing + 360

        # getting the rotation needed for our marker. We subtract 90 to account for the marker's orientation
        # of due East
        rotation = bearing - 90

        # get an evenly space list of lats and lons for our arrows. Note that I'm discarding the first and last for aesthetics
        # as I'm using markers to denote the start and end
        arrow_lats = np.linspace(p1.lat, p2.lat, n_arrows + 2)[1:n_arrows+1]
        arrow_lons = np.linspace(p1.lon, p2.lon, n_arrows + 2)[1:n_arrows+1]

        # Initialise the arrows marker coordinates list
        arrows = []

        #creating each "arrow" and appending them to our arrows list
        for points in zip(arrow_lats, arrow_lons):
            arrows.append(folium.RegularPolygonMarker(location=points, 
                          fill_color=color, number_of_sides=3, 
                          radius=size, rotation=rotation))
        return arrows
```

From the results above, there is far more time for people to be hiring bikes and taking round trips during the weekend than the weekdays because they most likely only work a 5-day week. In fact, there are more than twice the number of round trips on the weekend than during the week. The journey time during the weekend is longer too because people are less likely to be in a rush to get to places than during the week. The round trips during the week could suggest activity due to the constant hustle and bustle of tourists visiting Scotland's capital wanting to explore Edinburgh.

```python
# split cycle into weekdays and weekend 
cycleAB_weekdays = cycle_AB.loc[(cycle_AB.dow != 5) & (cycle_AB.dow != 6)]
cycleAB_weekend = cycle_AB.loc[(cycle_AB.dow == 5) | (cycle_AB.dow == 6)]

# Average number of round trips during week. We divide each length by (21) or (9) depending on if we want an average for 
# each weekdays and weekend day since there are 21 weekdays in September and 9 weekend days in September.
print('The average number of round trips during a weekday: {}'.format(int(len(cycleAB_weekdays)/(21))))
print('The average number of round trips per day during the weekend: {}\n'.format(int(len(cycleAB_weekend)/(9))))

# average journey time during weekend and weekdays for non-round trips in minutes (hence we divide by 60):
print('The average journey time for non-round trips during weekdays is: {} minutes.'.format(round(cycleAB_weekdays.duration.mean()/60)))
print('The average journey time for non-round trips during the weekend is: {} minutes.\n'.format(round(cycleAB_weekend.duration.mean()/60)))

# average journey distance during weekend and weekdays for non-round trips.
print('The average journey distance for non-round trips on weekdays is: {} km.'.format(round(cycleAB_weekdays.distance.mean(),2)))
print('The average journey distance for non-round trips on the weekend is: {} km.'.format(round(cycleAB_weekend.distance.mean(),2)))
```

The results above are threefold:

- Firstly, we expect more non-round trips to be taken during the week than round trips because of commuters travelling to work or making meetings etc. However, a striking difference between round and non-round average hire counts is that there are more trips during the weekend than during a typical day of the week for the former whereas for the latter, both weekend and weekday average hire count is roughly the same. This further supports the evidence that people are more likely to take a round-trip to explore Edinburgh during the weekend when they have more time to do so.


- Secondly, the average journey time for non-round trips during the week and during the weekend are almost half of the journey time taken for round trips which suggest that it just takes twice as long for an average round trip simply because the bike has been returned to its original starting station. Although, it could also suggest that people are more leisurely and take their time when it comes to doing round trips. This is likely because non-round trips are trips that have a designated objective at their destination such as biking to work, running an errand or meeting people across the city. Those people will not likely make their trip a round one because it would cost too much to keep the bike waiting at their destination for them to return home with. Instead, they will just hire another bike at the end of their errand.


- Finally, now that we have excluded the round trips from our journey distance calculation, we can see that people are travelling farther on non-round trips during the weekends by about 0.23km which is a large change from the valuesfound using the entire dataframe, 1.83km and 1.88km respectively. Now we have evidence to say that even those people taking non-round trips travel farther on the weekend than on the weekdays.

# Long and short trips (by duration)


We will no look at renting patterns based on the duration of the trips. Here we decided to use the median as a way of dividing the trips into short and long trips. The following code will plot the start(green) and end (red) stations for long and short trips on two different maps. The size of the cirlces is dependend on the frequency of usage of a station as either the starting or end point of a trip.

We have opted to do this analysis based on duration rather than distance because this way we can include data for round-trips. In addition, distance values are only based on the distance between start and end loction. This does not give any information about how far each cyclist cycled as we cannot determine which route the chose between the start and end point.

Furthermore we have not been provided with the longitude and latitude values for the stations but rather the values of where the bikes have been left. Because of this some stations have been defined with several longitude and latidtude values. We will therefore define the marker postions to be the mean of all values attributed the the unique stations.

```python
# predefine colours relating to different things in plots:
start_colour = 'green' # Start stations
end_colour = 'red' # end stations
long_colour = 'darkmagenta' # stations for long strips
short_colour = 'royalblue' # stations for short trips
```

```python
# Long trip visualisation
long_m = folium.Map(
    location=[55.962450, -3.176920],
    zoom_start=12
)

# define long cycles to be cycles with duration greater than the mean
long_cycles = cycle[cycle['duration'] > cycle['duration'].median()]

# define lists to store number of strips started and ended at each location. This is to determine the circle size.
N_start_trips = []
N_end_trips = []

# extract unique start location names
long_start_location_names = long_cycles['start_station_name'].unique()
long_start_locations = []

# for each station we define the location to be the average of all the longitude and latitude 
# values attributed to this station
for station in long_start_location_names:
    station_frame = Question2(long_cycles).trips_startend_station(station)[0]
    long_start_locations.append((np.mean(list(set(station_frame['start_station_latitude']))),
                           np.mean(list(set(station_frame['start_station_longitude'])))))

# extract unique end location names
long_end_location_names = long_cycles['end_station_name'].unique()
long_end_locations = []

# for each station we define the location to be the average of all the longitude and latitude 
# values attributed to this station
for station in long_end_location_names:
    station_frame = Question2(long_cycles).trips_startend_station(station)[1]
    long_end_locations.append((np.mean(list(set(station_frame['end_station_latitude']))),
                           np.mean(list(set(station_frame['end_station_longitude'])))))

    
# for every start and end station we count the number of trips started or ended and then we 
# add a circle with the corresponding radius and colour to the map:

for idx, station in enumerate(long_start_location_names):
    N_start_trips.append(len(Question2(long_cycles).trips_startend_station(station)[0]))
    folium.CircleMarker(fill_colour = start_colour, location = long_start_locations[idx], 
                        color = start_colour, fill=True,
                        radius=(N_start_trips[idx]/20)).add_to(long_m)
    
for idx, station in enumerate(long_end_location_names):
    N_end_trips.append(len(Question2(long_cycles).trips_startend_station(station)[1]))
    folium.CircleMarker(fill_colour = end_colour, location = long_end_locations[idx], 
                        color = end_colour, fill=True,
                        radius=(N_end_trips[idx]/20)).add_to(long_m)

# add the predefined legend to the map
macro = MacroElement()
macro._template = Template(template_start_end)
long_m.get_root().add_child(macro)

# show map
long_m
```

```python
# store the start hour for each trim
NewYork['hr'] = NewYork['started_at'].dt.hour.values

# extract the morning (5am -10am) and evening(4pm-9pm) data
NewYork_morning = NewYork.loc[(NewYork['hr']>5) & (NewYork['hr']<10)]
NewYork_evening = NewYork.loc[(NewYork['hr']>16) & (NewYork['hr']<21)]

# extract the start and end stations used for the morning trips
NewYork_morning_startend = pd.concat([NewYork_morning['start_station_name'], NewYork_morning['end_station_name']], ignore_index=True)
NewYork_morning_station_names=list(set(NewYork_morning_startend))

# define the station markers to be the mean of all longitude and latitude values associated with it:
NewYork_morning_marker_locations =[]
# loop through all stations that were used on the morning trips
for station in NewYork_morning_station_names:
    # store all start longitude and latitude values associated with the current station
    station_frame = Question2(NewYork).trips_startend_station(station)[0]
    # if there were no starting stations we store evening stations instead
    if station_frame.empty:
        # we store the endstation values 
        station_frame = Question2(NewYork).trips_startend_station(station)[1]
        # store the mean of all end station values
        NewYork_morning_marker_locations.append((np.mean(list(set(station_frame['end_station_latitude']))),
                           np.mean(list(set(station_frame['end_station_longitude'])))))
    else:
        # store the mean of all start station values
        NewYork_morning_marker_locations.append((np.mean(list(set(station_frame['start_station_latitude']))),
                           np.mean(list(set(station_frame['start_station_longitude'])))))


# extract the start and end stations used for the evening trips
NewYork_evening_startend = pd.concat([NewYork_evening['start_station_name'], NewYork_evening['end_station_name']], ignore_index=True)
NewYork_evening_station_names=list(set(NewYork_evening_startend))

# define the station markers to be the mean of all longitude and latitude values associated with it:
NewYork_evening_marker_locations =[]
# loop through all stations that were used on the morning trips
for station in NewYork_evening_station_names:
    # store all start longitude and latitude values associated with the current station
    station_frame = Question2(NewYork).trips_startend_station(station)[0]
    # if there were no starting stations we store evening stations instead
    if station_frame.empty:
        # we store the endstation values 
        station_frame = Question2(NewYork).trips_startend_station(station)[1]
         # store the mean of all end station values
        NewYork_evening_marker_locations.append((np.mean(list(set(station_frame['end_station_latitude']))),
                           np.mean(list(set(station_frame['end_station_longitude'])))))
    else:
        # store the mean of all start station values
        NewYork_evening_marker_locations.append((np.mean(list(set(station_frame['start_station_latitude']))),
                           np.mean(list(set(station_frame['start_station_longitude'])))))
```

# New York 2016-2019 trends and seasonal changes

In this section we show how the mean duration time and hire count relate to eachother over the years of 2016 to 2019. For this part of the analysis, there is a zipped folder *New York 2016-2019* that contained folders named *2016, 2017, 2018, 2019*. Within these folders there are spreadsheets that represent the cycling data over each month available. The data amounts to 8.3Gb and takes about 160 seconds to load this data into python notebook.

```python
# Define paths to access files.
newyork_path = 'New York 2016-2019/'
years = ['2016/', '2017/', '2018/', '2019/']

# Define dictionaries to store hire count and mean durations for each of the years
NY_trips = {'2016':[], '2017':[], '2018':[], '2019':[]}
NY_meanduration = {'2016':[], '2017':[], '2018':[], '2019':[]}
```

```python
import time as t
t0 = t.time()

# Loop over the number of years to process
for year in range(len(years)):
    print('Processing New York Data year: {}'.format(years[year]))
    for i, file in enumerate(os.listdir(newyork_path + years[year])):
        file = os.listdir(newyork_path + years[year])[i]
        print('Processing month {}'.format(i))
        
        # Load the .csv file
        data= pd.read_csv(newyork_path + years[year]+file)
        
        # Append the number of hire counts to the appropriate dictionary key year
        NY_trips['{}'.format(years[year].split('/')[0])].append(len(data))
        
        # Append the mean duration of hires to the appropriate dictionary key year
        try:
            NY_meanduration['{}'.format(years[year].split('/')[0])].append(data['tripduration'].mean())
        except:
            NY_meanduration['{}'.format(years[year].split('/')[0])].append(data['Trip Duration'].mean())
t1 = t.time()

print('Processing data took {} seconds'.format((t1-t0)))
```

```python
# Define the vector of time: We have 3 years of 12 months (2016-18) and 10 months in 2019
time_vector = [i for i in range(3*12 + 10)]

# Unpack the dictionary lists into a single list
NY_trip_vector = [NY_trips['{}'.format(years[year].split('/')[0])] for year in range(len(years))]
NY_trip_vector = [item for sublist in NY_trip_vector for item in sublist]

NY_meandur_vector = [NY_meanduration['{}'.format(years[year].split('/')[0])] for year in range(len(years))]
NY_meandur_vector = [item/60 for sublist in NY_meandur_vector for item in sublist]

# Determine correlation between the hire count and mean duration 
correlation = []
for i in range(len(NY_trip_vector)-2):
    correlation.append(np.corrcoef(NY_trip_vector[:i+2],NY_meandur_vector[:i+2])[0,1])

# Store the 3 results of data in a single list to iterate over for visualisation
variables = [NY_trip_vector, NY_meandur_vector, correlation]
```

```python
# Set up figure with 3 duplicate axes to plot 3 lines with 3 separate axis
fig = plt.figure(figsize = (18,8))
ax = fig.add_subplot(111)
ax.set_title('Mean Duration vs. Hire Count in New York 2016-2019', fontsize=20)
axes = [ax, ax.twinx(), ax.twinx()]

# Make some space on the right side for the extra y-axis.
fig.subplots_adjust(right=0.75)

# Move 2 of the y axes over to the right by 20% of the width of the axes
axes[-2].spines['right'].set_position(('axes', 1))
axes[-1].spines['right'].set_position(('axes', 1.1))

# Iteratively plot the 3 curves using colours and labels
colours = ['blue', 'red', 'green']
labels = ['Hire Count', 'Mean Duration', 'Correlation']
for idx, (ax, colour) in enumerate(zip(axes, colours)):
    data = variables[idx]
    if labels[idx] == 'Correlation':
        ax.plot(time_vector[2:], data, '-', color=colour, label = labels[idx])    
    else:
        ax.plot(time_vector, data, '-', color=colour, label = labels[idx])    
    ax.set_ylabel(labels[idx], color=colour, fontsize=18)
    ax.tick_params(axis='y', colors=colour)
axes[0].set_xlabel('Year', fontsize=18)
axes[0].set_xticklabels([years[i].split('/')[0] for i in range(len(years))])

# Set the year labels where the year starts since we have 7 months for 2013 and 10 months for 2019
ax.set_xticks([0, 12, 24, 36])
plt.show()

#Reason for draught in 2019: https://ny.curbed.com/2019/8/8/20726446/new-york-bike-lane-green-wave-analysis
```

The above plot shows how the hire count and mean duration for the bike sharing scheme develop over the years of 2016 to 2019, represented by the blue and red curves respectively. These two curves are evolving in a periodic fashion because of how periodic the weather can be across the years. The direct effect of weather and seasonal changes correspond to how many people choose to hire bikes and how long they choose to spend on them. This can be seen by looking at the middle regions of each of the yearly ticks, representing summertime, as well as the regions neighbouring the yearly ticks representing winter - the hire count and mean duration both increases at the same rate as well as decreases at the same rate, respectively. The green curve represents the correlation curve between these two variables, which for the most part shows a positive correlation coefficient of 0.5 and more, suggesting that throughout the yearly timeline of the bike sharing scheme there is a clear correlation between the mean duration time of a bike trip and the number of bike trips taken at the time of year. There is however a draught of hire counts in 2019 which is a suprise. However, looking at this [article](https://ny.curbed.com/2019/8/8/20726446/new-york-bike-lane-green-wave-analysis) written in August 2019, mentions that 

"As of this (August) month, 18 cyclists have been killed on city streets. That’s nearly double the fatalities suffered in all of 2018—and it’s only August."

Perhaps people don't want to risk taking to the streets using the bike scheme given the sudden rise in cyclist deaths, which may explain why there has been a sudden shortage in hire counts.

```python
# THIS IS THE SEASONAL CHANGE FOR NEW YORK FOR 2016-2019.

# Define months of the year and colours which represent the seasons
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November',
         'December']
colours = ['b','b','g', 'g','g','r','r','r', 'maroon','maroon', 'maroon', 'b']

# NOTE:     Blue: Winter     Green: Spring     Red: Summer     Marron: Autumn

# Plot 4 separate subplots to show seasonal changes 
fig = plt.figure(figsize=(20,14))
fig.suptitle('Mean Duration vs. Hire Count in New York', fontsize=20)
for axes in range(4):
    ax = fig.add_subplot(2,2,axes+1)
    year = years[axes].split('/')[0]
    for i in range(len(NY_meanduration['{}'.format(year)])):
        x = NY_meanduration['{}'.format(year)][i]/60
        y = NY_trips['{}'.format(year)][i]
        ax.plot(x,y, label = months[i], marker='*', markersize=15, color = colours[i])
        ax.set_title('{}'.format(year), fontsize=20)
        ax.set_xlabel('Mean Duration (minutes)', fontsize=18)
        ax.set_ylabel('Hire Count', fontsize=18)  
        
        # Add a 
        corr = 0.15
        ax.annotate(str(months[i][:3]),  xy = (x + corr, y - corr), fontsize=14)
        
        # There was a deficit in hire counts in 2019, hence why the y axis has been limited differently
        if year != '2019':
            ax.set_ylim([0, 2000000])
        else:
            ax.set_ylim([0, 55000])
```

The 4 subplots above show the seasonal change in mean duration and hire count for the 4 years of New York bike scheme data. There is a clear region of winter (blue) and summer (red) data points that show for hotter months we expect more and longer bike trips and colder months the opposite. The months of spring and autumn data points exist in the middle region between summer and winter, which further suggest the relationship between hire count and mean duration with respect to particular times of the year.
